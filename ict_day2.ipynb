```python
q1=np.percentile(datas['spirit_servings'],25)
q3=np.percentile(datas['spirit_servings'],75)
q2=np.percentile(datas['spirit_servings'],50)
iqr=q3-q1
print(q1,q2,q3,iqr)
```

```python
datas['spirit_servings'].median()
```

```python
up_limt= q3+1.5*iqr
l_limit= q1-1.5*iqr
print(up_limt,l_limit)
```

```python
outlier= []
for x in datas['spirit_servings']:
  if (x>up_limt) or (x<l_limit):
    outlier.append(x)
print(outlier)
```

```python
q1=np.percentile(datas['wine_servings'],25)
q3=np.percentile(datas['wine_servings'],75)
q2=np.percentile(datas['wine_servings'],50)
iqr=q3-q1
print(q1,q2,q3,iqr)
```

```python
datas['wine_servings'].median()
```

```python
outlier= []
q1=np.percentile(datas['wine_servings'],25)
q3=np.percentile(datas['wine_servings'],75)
q2=np.percentile(datas['wine_servings'],50)
iqr=q3-q1
print(q1,q2,q3,iqr)
up_limt= q3+1.5*iqr
l_limit= q1-1.5*iqr
print(up_limt,l_limit)
for x in datas['wine_servings']:
  if (x>up_limt) or (x<l_limit):
    outlier.append(x)
print(outlier)
```
