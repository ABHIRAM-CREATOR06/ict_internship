STRING INDEXING/SLICING

```python
s='python'
print(s[0])
print(len(s)) #STARTS WITH 0 TO LEN-1
print(s[-1])  #starts with -1 ends with - len reverse order
print(s[1:4])
print(s[:5])  #slices from 1 to 5 indexes
print(s[2:])  #slices from 2 to len-1
print(s[-5:-2])
print(s[1:5:2]) #prints index 1 then ignores every alternativing index

```
