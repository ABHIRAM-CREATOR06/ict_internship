```python
#question 6
n=int(input("enter limit: "))
if n<2:
  print("not prime")
else:
  for i in range(2,n):
    if n%i==0:
      print("not prime")
      break
  else:
    print("prime")
```

```python
#question 7

numbers = [10, 40, 30, 70]

print("Original List :", numbers)

reversed_list = []
for i in range(len(numbers) - 1, -1, -1):
    reversed_list.append(numbers[i])

print("Reversed List :", reversed_list)
```
